```python
import requests
from bs4 import BeautifulSoup
import re
import os
from requests.exceptions import RequestException

headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/66.0.3359.181 Safari/537.36'
    }

#DOWNLOAD_FOLDER
DOWNLOAD_FOLDER = os.path.abspath('images')
if not os.path.exists(DOWNLOAD_FOLDER):
    os.mkdir(DOWNLOAD_FOLDER)

def get_page_detail(item):
    url = item.get('url')
    try:
        response = requests.get(url,headers = headers)
        if response.status_code == 200:
            return response.text

    except RequestException:
        print('Solicitud de página web fallida')
        return None
    
def get_page():
    
    url = 'https://www.dafiti.com.ar/combo-zapatillas-2022/'
    response = requests.get(url,headers = headers)
    if response.status_code == 200:
        return response.text



def parse_page(html):
    soup = BeautifulSoup(html,'html.parser')
    
    imgs = soup.find_all(name = 'img', class_='itm-img')
    
    for img in imgs:
        yield{

        'title': img.get('alt'),

        'url':img.get('data-src')

        }
    
def save_pic(pic):
    title = pic.get('title')

    #eliminar símbolos, caracteres ilegales | etc.
    title = re.sub('[\/:*?"<>|]','-',title).strip()
    url = pic.get('url')
    
        
    response = requests.get(url,headers = headers)
    try:
        if response.status_code == 200:
            aux_file_path = '{0}.{1}' .format(title,'jpg')
            
            
            file_path = os.path.join(DOWNLOAD_FOLDER, aux_file_path)
            if not os.path.exists(file_path):
                with open(file_path,'wb') as f:
                    f.write(response.content)
                    print('Se ha descargado la imagen {0}' .format(title))
            else:
                print('La imagen {0}, ya ha sido descargada'.format(title))
                 

    except RequestException as e:
        print(e)
        return None
    
def main():
    """
    main function
    """
    html = get_page()
    data = parse_page(html)  
    for item in data:
        save_pic(item)
```

```python
if __name__ == '__main__':
    main()
```
